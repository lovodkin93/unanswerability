```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# T5

```python
# Create a DataFrame from the new data
data = {'SQuAD': [89.9, 84.1, 71.5],
        'NQ': [74.9, 86.1, 65.5],
        'MuSiQue': [75, 76.1, 77.5]}
df = pd.DataFrame(data, index=['SQuAD', 'NQ', 'MuSiQue'])

# Create a heatmap
plt.figure(figsize=(8, 6))
ax = sns.heatmap(df, annot=True, fmt=".1f", cmap='YlGnBu', annot_kws={"size": 20}, vmin=16, vmax=91)  # Adjust scale here

plt.tick_params(axis='both', which='major', labelsize=15)

# Get the colorbar from the heatmap
cbar = ax.collections[0].colorbar

# Here you can set the fontsize of colorbar tick labels
cbar.ax.tick_params(labelsize=16) 

plt.show()
```

# Shared

```python
# Define your data
data1 = {'SQuAD': [89.9, 84.1, 71.5],
        'NQ': [74.9, 86.1, 65.5],
        'MuSiQue': [75, 76.1, 77.5]}

data2 = {'SQuAD': [90.4, 82.4, 73.4],
        'NQ': [85.6, 87.3, 73.5],
        'MuSiQue': [64.7, 82.5, 78.3]}

data3 = {'SQuAD': [82.8, 84.4, 63.6],
        'NQ': [78, 85.5, 60],
        'MuSiQue': [17.4, 41.5, 75.5]}

df1 = pd.DataFrame(data1, index=['SQuAD', 'NQ', 'MuSiQue'])
df2 = pd.DataFrame(data2, index=['SQuAD', 'NQ', 'MuSiQue'])
df3 = pd.DataFrame(data3, index=['SQuAD', 'NQ', 'MuSiQue'])

# Find global min and max
vmin = np.min([df1.values.min(), df2.values.min(), df3.values.min()])
vmax = np.max([df1.values.max(), df2.values.max(), df3.values.max()])

# Create a 3x1 subplot
fig, axs = plt.subplots(3, 1, figsize=(8, 18), sharex=True)

# Create colorbar
cbar_ax = fig.add_axes([.91, .1, .03, .8])  # coordinates changed here

# Create heatmaps
sns.heatmap(df1, annot=True, fmt=".1f", cmap='YlGnBu', annot_kws={"size": 16}, ax=axs[0], cbar=False, vmin=vmin, vmax=vmax)
sns.heatmap(df2, annot=True, fmt=".1f", cmap='YlGnBu', annot_kws={"size": 16}, ax=axs[1], cbar=False, vmin=vmin, vmax=vmax)
sns.heatmap(df3, annot=True, fmt=".1f", cmap='YlGnBu', annot_kws={"size": 16}, ax=axs[2], cbar=True, cbar_ax=cbar_ax, vmin=vmin, vmax=vmax)

# Set the fontsize of colorbar tick labels
cbar = axs[2].collections[0].colorbar
cbar.ax.tick_params(labelsize=16) 

# Add captions
captions = ["T5", "UL2", "OPT"]
for ax, cap in zip(axs, captions):
    ax.set_xlabel(cap, size=18)

plt.show()
```
