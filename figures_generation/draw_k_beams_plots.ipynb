```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

# Squad

```python
# Assuming your data is in the following format
data = {
    "model": ["Flan-T5$_{xxl}$", "Flan-T5$_{xxl}$", "Flan-T5$_{xxl}$", "Flan-T5$_{xxl}$", "Flan-UL2", "Flan-UL2", "Flan-UL2", "Flan-UL2", "OPT-IML", "OPT-IML", "OPT-IML", "OPT-IML"],
    "# beams": [1, 3, 5, 7, 1, 3, 5, 7, 1, 3, 5, 7],
    "any beam": [37.3, 53.3, 62.3, 67.5, 46.3, 65.4, 72.8, 76.2, 43.9, 55.3, 62.3, 65.2],
    "top beam": [37.3, 38.9, 38.1, 37.8, 46.3, 49.6, 50.1, 50.0, 43.9, 40.8, 40.8, 41.1]
}

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Define figure size
plt.figure(figsize=(10,6))


# Set larger font sizes
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20
mpl.rcParams['legend.fontsize'] = 13


# Create bar plot
barplot = sns.barplot(data=df, x='model', y='any beam', hue='# beams')

# Adjust width of bars to ensure the horizontal lines can fit within the bars
for bar in barplot.containers:
    for patch in bar.patches:
        current_width = patch.get_width()
        diff = current_width - .2

        # we change the bar width
        patch.set_width(.2)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

# Add a horizontal line inside each bar for the top beam score
for i, model in enumerate(df['model'].unique()):
    df_model = df[df['model'] == model]
    for j, row in enumerate(df_model.iterrows()):
        bar_center = i + (j - len(df_model)/2 + 0.5) * 0.2
        plt.hlines(row[1]['top beam'], bar_center-0.1, bar_center+0.1, colors='k')

# Remove x-axis title
barplot.set_xlabel('')

plt.ylabel('F1 (unanswerable)')
# plt.legend(title='# beams')

# Increase legend title font size
legend = barplot.get_legend()
legend.set_title('beam size', prop={'size': 13})


plt.show()
```
